Na modalidade de juros simples, o cálculo da taxa de juros é feito pela fórmula i=(m- c)/(c.n) onde "i" é a taxa de juros simples, "c" é o capital, "m" é montante e "n" é o tempo. Se uma fatura no valor de R$ 5.387,76 foi paga com 23 dias de atraso totalizando R$ 5.612,89, qual foi a taxa diária de juros utilizada?

```python
C=5387.76
M=5612.89
N=23
I=(M-C)/(C*N)*100
print('A Taxa de juros em porcentagem foi %.2f%% ao dia'%I)
```

Na modalidade de juros compostos, o cálculo do valor futuro é feito pela fórmula vf=vp*(1+i)^n, onde "vf" é o valor futuro, "vp" é o valor presente, "i" é a taxa de crescimento e "n" é o tempo. Qual o valor futuro de uma aplicação de R$ 12.300,00 feita por 10 meses a uma taxa de juros compostos de 2% ao mês?

```python
C=12300
I=2/100
N=10
M=C*(1+I)**N
print('O Valor Futuro da Aplicação é de: R$%.2f' %M)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

x= symbols("x")


# Definindo os pontos
x1, y1 = 3, 2
x2, y2 = 6, 3

# Calculando o coeficiente angular (m) e o coeficiente linear (b)
m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1
y = m * x + b
# Exibindo a equação da reta
print(f"A equação da reta que interpola os pontos ({x1}, {y1}) e ({x2}, {y2}) é: y = {m}x + {b}")

# Criando os valores de x para o gráfico
x_values = np.linspace(2, 7, 100)  # Valores de x entre 2 e 7
y_values = m * x_values + b  # Calculando os valores de y

# Criando o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x_values, y_values, label=f'y = {m:.2f}x + {b:.2f}', color='blue')  # Linha da reta
plt.scatter([x1, x2], [y1, y2], color='red')  # Pontos dados
plt.text(x1, y1, f'({x1}, {y1})', fontsize=12, verticalalignment='bottom', horizontalalignment='right')
plt.text(x2, y2, f'({x2}, {y2})', fontsize=12, verticalalignment='bottom', horizontalalignment='right')

# Configurações do gráfico
plt.title('Gráfico da Reta Interpoladora')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.legend()
plt.xlim(2, 7)
plt.ylim(0, 5)

# Exibindo o gráfico
plt.show()
y1=y.subs(x,7)
print(y1)
```

```python
from sympy import *

# Definindo a variável simbólica
x = symbols('x')

# Definindo a função
f = -4*x**2 + 40*x - 7

# Encontrando os pontos de interseção com o eixo x
pontos_interseccao = solve(Eq(f, 0), x)
print(f"Pontos de interseção com o eixo x: {pontos_interseccao}")

# Calculando a área entre a curva e o eixo x
# A área é dada pela integral da função entre os pontos de interseção
area = integrate(f, (x, pontos_interseccao[0], pontos_interseccao[1]))
print(f"A área limitada pelo gráfico da função e pelo eixo x é: {area}")
```

```python
import numpy as np

# Definindo a matriz dos coeficientes
A = np.array([[5, 1, 3],
              [-1, 2, 5],
              [4, -5, 1]])

# Definindo o vetor dos termos independentes
b = np.array([76, 35, 22])

# Resolvendo o sistema linear usando a função `linalg.solve`
x = np.linalg.solve(A, b)

# Imprimindo a solução
print("A solução do sistema linear é:")
print(f"x = {x[0]}")
print(f"y = {x[1]}")
print(f"z = {x[2]}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definindo os pontos
x = np.array([3, 4, 7, 10])
y = np.array([6, 4, 11, 12])

# Encontrando os coeficientes do polinômio
coeficientes = np.polyfit(x, y, 3)  # grau 2 para um polinômio quadrático

# Criando a função polinomial
polinomio = np.poly1d(coeficientes)

# Exibindo a função
print(f"A função do polinômio é: {polinomio}")

# Para visualizar o polinômio
x_vals = np.linspace(-2, 15, 9)
y_vals = polinomio(x_vals)

plt.plot(x_vals, y_vals, label='Polinômio Interpolador')
plt.scatter(x, y, color='red', label='Pontos Originais')
plt.title('Interpolação de Polinômios')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.grid()
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a função de custo
def custo(x):
  return 2*x**2 - 40*x + 1000

# Cálculo da derivada da função de custo
def derivada_custo(x):
  return 4*x - 40

# Encontrar o ponto crítico (onde a derivada é zero)
x_critico = -derivada_custo(0) / 4  # Resolvendo 4x - 40 = 0
print(f"Ponto crítico: x = {x_critico}")

# Verificar se o ponto crítico é um mínimo
if derivada_custo(x_critico - 0.1) > 0 and derivada_custo(x_critico + 0.1) < 0:
  print("O ponto crítico é um mínimo.")
else:
  print("O ponto crítico não é um mínimo.")

# Plotar a função de custo e sua derivada
x = np.linspace(0, 20, 100)
y_custo = custo(x)
y_derivada = derivada_custo(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y_custo, label="Custo")
plt.plot(x, y_derivada, label="Derivada do Custo")
plt.xlabel("Nível de Produção (x)")
plt.ylabel("Custo / Derivada")
plt.title("Custo e Derivada do Custo")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Instalando o PuLP se necessário
try:
    import pulp
except ImportError:
    !pip install pulp

# Importando a biblioteca PuLP
from pulp import LpProblem, LpMaximize, LpVariable, lpSum, LpStatus, value

# Criando o problema
problema = LpProblem("Maximizar_Lucro", LpMaximize)

# Definindo as variáveis
x = LpVariable('x', lowBound=0)
x2 = LpVariable('x2', lowBound=0)
x3 = LpVariable('x3', lowBound=0)

# Definindo a função objetivo
problema += 3*x+4*x2+2*x3, "Lucro_Total"

# Definindo as restrições
problema += x+10*x2+4*x3<=120
problema += 4*x+x2<=80
problema += 2*x2+3*x3<=78

for i in range(10):
  problema += (x ==i) >> (x2 == i * i)
  problema += (x ==i) >> (x2 == i * i * i)

# Resolvendo o problema
problema.solve()

# Resultados
print("Lucro total máximo: R$", value(problema.objective))
```

```python
from pulp import *

# Cria um problema de maximização
problema = LpProblem("Problema_Linear", LpMaximize)

# Define as variáveis de decisão
x1 = LpVariable("x1", lowBound=0)
x2 = LpVariable("x2", lowBound=0)
x3 = LpVariable("x3", lowBound=0)

# Define a função objetivo
problema += 3*x1 + 4*x2 + 2*x3, "Função Objetivo"

# Define as restrições
problema += x1 + 10*x2 + 4*x3 <= 120, "Restrição 1"
problema += 4*x1 + x2 <= 80, "Restrição 2"
problema += 2*x2 + 3*x3 <= 78, "Restrição 3"

# Resolve o problema
problema.solve()

# Imprime o resultado
print("Status do problema:", LpStatus[problema.status])

if problema.status == LpStatusOptimal:
    print("Valor máximo da função objetivo:", value(problema.objective))
    print("Valores das variáveis:")
    print("x1 =", value(x1))
    print("x2 =", value(x2))
    print("x3 =", value(x3))
```
